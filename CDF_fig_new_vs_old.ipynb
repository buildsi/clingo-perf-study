```python
import pandas as pd
import numpy as np
import glob
import re
import matplotlib.pyplot as plt
```

```python
machine = 'lassen'
df = pd.read_csv('{0}/out_log_full_complete.csv'.format(machine), header=None, names=['pkg', 'cfg', 'iter', 'setup', 'load', 'ground', 'solve', 'total', 'dep_len'])
df_full = df
```

```python
df_old = pd.read_csv('{0}/log_old_concretizer.csv'.format(machine), header=None, names=['pkg', 'iter', 'success', 'total'])
df_old_filtered = df_old[df_old['iter'] == 0]
```

```python
packs_old = df_old_filtered['pkg']
pk_ls_old = sorted(list(set(packs_old)))
print(len(pk_ls_old))
```

```python
packs = df_full['pkg']
pk_ls = sorted(list(set(packs)))
print(len(pk_ls))
```

```python
confs = df_full['cfg']
cfg_ls = list(set(confs))
print(cfg_ls)
```

```python
phases = ['setup', 'load', 'ground', 'solve', 'total']
```

```python
timings = {}

for cf in cfg_ls:
    timings[cf] = {}
    for ph in phases:
        timings[cf][ph] = []
    for pk in pk_ls:
        req_pk = (df_full['pkg'] == pk)
        req_cf = (df_full['cfg'] == cf)
        tdf = df_full[req_pk & req_cf]
        for ph in phases:
            timings[cf][ph].append(tdf[ph].median())
```

```python
old_timings = []

for pk in pk_ls_old:
    req_pk = (df_old_filtered['pkg'] == pk)
    tdf = df_old_filtered[req_pk]
    old_timings.append(tdf['total'].median())
```

```python
solve_times = sorted(zip(pk_ls_old, old_timings), key=lambda x: x[1], reverse=True)
solve_times[0:20]
```

```python
for i, ph in enumerate(phases):
    fig = plt.figure(figsize=(6, 6), dpi=150)
    ax = fig.add_subplot(1, 1, 1)

    n_bins = 500

#     cfgs = [' tweety', ' trendy', ' handy']
    cfgs = [' tweety', 'old']
    colors = ['red', 'blue', 'magenta', 'black', 'black']

    fntsz = 20
#     ph = phases[0]

    for j, cf in enumerate(cfgs):
        if cf == 'old':
            solve_times = sorted(zip(pk_ls_old, old_timings), key=lambda x: x[1], reverse=True)
            tab_data = [[p, "{:.3f}".format(t)] for p, t in solve_times[0:5]]
            arr = np.array(sorted(old_timings, reverse=True))
            values, base = np.histogram(arr, bins=n_bins)
            cumulative = np.cumsum(values)
            ax.plot(base[:-1], cumulative, c=colors[j], label='Old concretizer')
        else:
            solve_times = sorted(zip(pk_ls, timings[cf][ph]), key=lambda x: x[1], reverse=True)
            tab_data = [[p, "{:.3f}".format(t)] for p, t in solve_times[0:5]]
            arr = np.array(sorted(timings[cf][ph], reverse=True))
            values, base = np.histogram(arr, bins=n_bins)
            cumulative = np.cumsum(values)
            ax.plot(base[:-1], cumulative, c=colors[j], label='Clingo')
        ax.legend(fontsize='x-large')
        ax.set_xlabel('Time [s]', fontsize=fntsz)
        ax.set_ylabel('Package count', fontsize=fntsz)
        ax.set_yticks([0, 1000, 2000, 3000, 4000, 5000, 6000])
        ax.set_yticklabels(['0', '1K', '2K', '3K', '4K', '5K', '6K'])
        ax.tick_params(axis='both', labelsize=fntsz)
    #     ax.title.set_text(ph)

    # plt.show()
    plt.savefig('cdf_{0}_{1}_fig_n_on.pdf'.format(machine, ph), dpi=150, bbox_inches='tight')
```
