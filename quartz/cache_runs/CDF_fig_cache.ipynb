```python
machine = 'lassen'
df_cache = []
df_cache.append(pd.read_csv('../../../{0}/cache_runs/out_log_cache_1.csv'.format(machine), header=None, names=['pkg', 'cfg', 'iter', 'setup', 'load', 'ground', 'solve', 'total', 'dep_len']))
df_cache.append(pd.read_csv('../../../{0}/cache_runs/out_log_cache_2.csv'.format(machine), header=None, names=['pkg', 'cfg', 'iter', 'setup', 'load', 'ground', 'solve', 'total', 'dep_len']))
df_cache.append(pd.read_csv('../../../{0}/cache_runs/out_log_cache_3.csv'.format(machine), header=None, names=['pkg', 'cfg', 'iter', 'setup', 'load', 'ground', 'solve', 'total', 'dep_len']))
df_cache.append(pd.read_csv('../../../{0}/cache_runs/out_log_cache_full.csv'.format(machine), header=None, names=['pkg', 'cfg', 'iter', 'setup', 'load', 'ground', 'solve', 'total', 'dep_len']))
```

```python
packs = df_cache[0]['pkg']
pk_ls = sorted(list(set(packs)))
print(len(pk_ls))
```

```python
packs = set(limited_packs)
for df in df_cache:
    packs = packs.intersection(set(df['pkg']))
pk_ls = sorted(list(set(packs)))
```

```python
print(len(pk_ls))
```

```python
confs = df_cache[0]['cfg']
cfg_ls = list(set(confs))
print(cfg_ls)
```

```python
df = df_cache[0]
pk_ls[0]
```

```python
req_pk = df['pkg'] == pk_ls[0]
req_cf = df['cfg'] == cfg_ls[0]
df[req_pk & req_cf]
```

```python
phases = ['setup', 'load', 'ground', 'solve', 'total']
```

```python
timings_list = []

for df in df_cache:
    timings_list.append({})
    curr_timings = timings_list[-1]
    for cf in cfg_ls:
        curr_timings[cf] = {}
        for ph in phases:
            curr_timings[cf][ph] = []
        for pk in pk_ls:
            req_pk = df['pkg'] == pk
            req_cf = df['cfg'] == cf
            tdf = df[req_pk & req_cf]
            for ph in phases:
                curr_timings[cf][ph].append(tdf[ph].median())
```

```python
min(timings_list[0][cfg_ls[0]]['total'])
```

```python
cfgs = [' tweety']
colors = ['blue', 'red', 'magenta', 'black']
labels = ['6804 cached pkgs', '15255 cached pkgs', '27160 cached pkgs', '63099 cached pkgs']

for i, ph in enumerate(phases):
    fig = plt.figure(figsize=(6, 6), dpi=150)
    ax = fig.add_subplot(1, 1, 1)

    n_bins = 500
    fntsz = 20

    for i_tm, tms in enumerate(timings_list):
        for cf in cfgs:
            solve_times = sorted(zip(pk_ls, tms[cf][ph]), key=lambda x: x[1], reverse=True)
            tab_data = [[p, "{:.3f}".format(t)] for p, t in solve_times[0:5]]
            arr = np.array(sorted(tms[cf][ph], reverse=True))
            values, base = np.histogram(arr, bins=n_bins)
            cumulative = np.cumsum(values)
            ax.plot(base[:-1], cumulative, c=colors[i_tm], label=labels[i_tm])
            ax.legend(fontsize='x-large')
            ax.set_xlabel('Time [s]', fontsize=fntsz)
            ax.set_ylabel('Package count', fontsize=fntsz)
            ax.set_yticks([0, 100, 200, 300, 400, 500, 600])
            ax.tick_params(axis='both', labelsize=fntsz)
            
    plt.savefig('cdf_e4s_cache_{0}_{1}_fig_n.pdf'.format(machine, ph), dpi=150, bbox_inches='tight')
```
